### Traversing tuple

````python
t=('a','b','c','d')
for ele in t:
    print(ele,end=" ")

print("\n")

for i in range(len(t)):
    print(t[i],end=" ")


#Time complexity O(n) space O(1)
````

## ðŸ§© Tuple vs List in Python

| Feature                           | **Tuple**                                                   | **List**                                                  |
| --------------------------------- | ----------------------------------------------------------- | --------------------------------------------------------- |
| **Syntax**                        | `(1, 2, 3)`                                                 | `[1, 2, 3]`                                               |
| **Mutability**                    | âŒ Immutable (cannot change after creation)                  | âœ… Mutable (can add, remove, or change items)              |
| **Methods available**             | Few (`count()`, `index()`)                                  | Many (`append()`, `extend()`, `remove()`, `sort()`, etc.) |
| **Speed**                         | âš¡ Faster (fixed size)                                       | ðŸ¢ Slower (dynamic resizing)                              |
| **Memory usage**                  | Less                                                        | More                                                      |
| **Hashable (usable as dict key)** | âœ… Yes (if all elements are immutable)                       | âŒ No                                                      |
| **Iteration**                     | Slightly faster                                             | Slightly slower                                           |
| **Use case**                      | Fixed collections of items (e.g., coordinates, config data) | Collections that change frequently                        |
| **Example modification**          | âŒ `t[0] = 5` â†’ Error                                        | âœ… `lst[0] = 5` â†’ Works                                    |
| **Creation**                      | `t = (1, 2, 3)` or `tuple([1,2,3])`                         | `lst = [1, 2, 3]` or `list((1,2,3))`                      |

### ðŸ’¬ Common Interview Question Tips

**Q1:** Why are tuples faster than lists?

> Tuples are stored in a fixed-size memory block, while lists need dynamic resizing and extra references for mutability.

**Q2:** Can a tuple be a dictionary key?

> âœ… Yes, if all its elements are immutable.

**Q3:** When should you use a tuple?

> When your data is fixed and should not change (e.g., coordinates, config, constants).
