### ARRAY vs LINKED LIST

| Feature                      | **Array**                                              | **Linked List**                                                              |
| ---------------------------- | ------------------------------------------------------ | ---------------------------------------------------------------------------- |
| **Definition**               | Collection of elements stored **in contiguous memory** | Collection of **nodes**, each containing data and a pointer to the next node |
| **Memory allocation**        | **Fixed size** (static)                                | **Dynamic size** (can grow or shrink easily)                                 |
| **Access time**              | **O(1)** (direct indexing via index)                   | **O(n)** (must traverse from head)                                           |
| **Insertion/Deletion**       | Slow â€” may require shifting elements (**O(n)**)        | Fast â€” just change pointers (**O(1)** at head)                               |
| **Memory usage**             | Compact (only data stored)                             | Extra memory for pointer/reference                                           |
| **Cache locality**           | Better (contiguous memory)                             | Poor (nodes may be scattered)                                                |
| **Implementation in Python** | Using `list` or `array` module                         | Implemented via classes and nodes                                            |
| **Example**                  | `[10, 20, 30, 40]`                                     | `10 â†’ 20 â†’ 30 â†’ 40 â†’ None`                                                   |



## ðŸ”— **Types of Linked Lists in Python / DSA**

Linked Lists are **linear data structures** where elements (called **nodes**) are connected using **pointers/references**.
Each node contains **data** and one or more **links** to other nodes.

---

### ðŸŸ¢ **1ï¸âƒ£ Singly Linked List**

**ðŸ“˜ Structure:**
Each node stores:

* `data`
* a reference (`next`) to the next node.

**ðŸ§© Visual Representation:**

```
Head â†’ [10 | Next] â†’ [20 | Next] â†’ [30 | None]
```

**ðŸ’¡ Explanation:**

* The `Head` points to the first node.
* Each node points to the next one.
* The last node points to `None` (end of list).

**âž¡ï¸ Traversal:** Only **forward** direction.

---

### ðŸ”µ **2ï¸âƒ£ Circular Singly Linked List**

**ðŸ“˜ Structure:**
Similar to singly linked list, but the **last node points back to the first node**.

**ðŸ§© Visual Representation:**

```
Head â†’ [10 | Next] â†’ [20 | Next] â†’ [30 | Next] --\
        ^------------------------------------------/
```

**ðŸ’¡ Explanation:**

* No node points to `None`.
* The last nodeâ€™s `next` links back to the `Head`.
* You can start at any node and keep moving forever.

**âž¡ï¸ Traversal:** **Forward only**, but continuous (circular).

---

### ðŸŸ  **3ï¸âƒ£ Doubly Linked List**

**ðŸ“˜ Structure:**
Each node stores:

* `data`
* pointer to **previous node (`prev`)**
* pointer to **next node (`next`)**

**ðŸ§© Visual Representation:**

```
None â† [10 | Prev | Next] â†” [20 | Prev | Next] â†” [30 | Prev | Next] â†’ None
```

**ðŸ’¡ Explanation:**

* Each node is connected **both ways**.
* You can move **forward** or **backward** in the list.

**âž¡ï¸ Traversal:** **Bidirectional** (both directions).

---

### ðŸ”´ **4ï¸âƒ£ Circular Doubly Linked List**

**ðŸ“˜ Structure:**
A **doubly linked list** where the **last node connects back to the first node** and vice versa.

**ðŸ§© Visual Representation:**

```
Head â†” [10 | Prev | Next] â†” [20 | Prev | Next] â†” [30 | Prev | Next] â†” Head
```

**ðŸ’¡ Explanation:**

* The `Head`â€™s `prev` points to the last node.
* The last nodeâ€™s `next` points to the `Head`.
* You can move **forward or backward infinitely**.

**âž¡ï¸ Traversal:** **Circular + Bidirectional**

---

### ðŸ§  **Summary Table**

| Type                        | Next Pointer | Prev Pointer | Circular? | Traversal Direction           |
| --------------------------- | ------------ | ------------ | --------- | ----------------------------- |
| Singly Linked List          | âœ…            | âŒ            | âŒ         | Forward only                  |
| Circular Singly Linked List | âœ…            | âŒ            | âœ…         | Forward only (looped)         |
| Doubly Linked List          | âœ…            | âœ…            | âŒ         | Forward & backward            |
| Circular Doubly Linked List | âœ…            | âœ…            | âœ…         | Forward & backward infinitely |

---

### âœ… **In short:**

> * **Singly LL:** Linear â†’ One-way.
> * **Circular Singly LL:** Linear â†’ Loops to start.
> * **Doubly LL:** Two-way connections.
> * **Circular Doubly LL:** Two-way + circular traversal.

````python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None

class Linkedlist:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0

    def append(self,value):
        if self.head is None:
            new_node=Node(value)
            self.head=new_node
            self.tail=new_node
        else:
            new_node=Node(value)
            self.tail.next=new_node
            self.tail=new_node

        self.length+=1

    
````
