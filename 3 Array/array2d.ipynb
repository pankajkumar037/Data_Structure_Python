### Insertion in an Array

```python
arr=np.array([[1,2,3],[5,6,7]])
```

```python
new_arr=np.insert(arr,1,[[0,0,0]],axis=0)
#array object,Position of that, elements to be inserted,axis
```

```python
print(arr)
print(new_arr)
```

```python
#adding to end of 2d aray
new_arr=np.append(arr,[[0,0,0]],axis=0)
print(arr)
print(new_arr)
```

### Array traversal

```python
import numpy as np

array=np.array([[1,2,3],[5,6,7]])
```

```python
for i in range(len(array)):
    for j in range(len(array[0])):
        print(array[i][j],end= " ")
    print("")
    
```

### deleting element in 2d Array

```python
import numpy as np

array=np.array([[1,2,3],[5,6,7]])
```

```python
del_arr=np.delete(array,0,axis=0)
```

```python
print(array)
print(del_arr)
```
