```python
li=[1,2,3,5,-1]
new_lst=[l*2 for l in li if l>0]
new_lst
```

```python
#The method isalpha() checks whether a string contains only alphabetic characters (A–Z or a–z).
```

```python
def is_consonat(letter):
    vowels="aeiou"
    return letter.isalpha() and letter.lower() not in vowels
```

```python
li=['a','b','c','d','e']
new_lst=[l for l in li if is_consonat(l)]
new_lst
```

```python
li=[1,2,3,5,-1]
new_lst=[l if l>0 else "negative" for l in li]
new_lst
```

```python
li=[1,2,3,5,-1]
def g_n(l):
    if l>0 :
        return l
    else:
        return "negative" 
new_lst=[g_n(l) for l in li]
new_lst
```
