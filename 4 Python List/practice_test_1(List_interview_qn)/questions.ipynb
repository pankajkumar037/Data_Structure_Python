```python
a=[1,2,3,4,5]
print(a[3:0:-1])
```

```python
"""
The output of the code is:

[4, 3, 2]
This is because of how list slicing works in Python. When you slice a list, you specify a start index, a stop index, and a step. The format is my_list[start:stop:step].

In your example:

start is 3 (the value at this index is 4).

stop is 0 (the slice will stop before reaching this index).

step is -1 (this tells Python to slice the list in reverse).

The slice starts at index 3 and moves backward one step at a time, collecting the elements at indices 3, 2, and 1. It stops just before reaching index 0, resulting in the list [4, 3, 2].
"""
```

```python
# What will be the output of the following code block?

fruit_list1 = ['Apple', 'Berry', 'Cherry', 'Papaya']
fruit_list2 = fruit_list1
fruit_list3 = fruit_list1[:]
 
fruit_list2[0] = 'Guava'
fruit_list3[1] = 'Kiwi'
 
sum = 0
for ls in (fruit_list1, fruit_list2, fruit_list3):
    if ls[0] == 'Guava':
        sum += 1
    if ls[1] == 'Kiwi':
        sum += 20
 
print(sum)
```
