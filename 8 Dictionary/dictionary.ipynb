### Search an element in Dictionary

```python
#by linear search
my_dict={"name":"Pankaj","age":27,"address":"Bhihar"}

def search(d,value):
    for key in d:
        if d[key] == value:
            return key,value
    
    return "not found"

print(search(my_dict,27))
print(search(my_dict,100))
```

```python
#Time Complexity=O(n)
#Space complexity=O(1)
```

| Operation                   | Description                                         | Average Time | Worst Time |

| `value in dict.values()`    | Check if value exists in dictionary                 | **O(n)**     | **O(n)**   |
| `(k, v) in dict.items()`    | Check if key-value pair exists                      | **O(n)**     | **O(n)**   |
| `dict[key]`                 | Access value by key                                 | **O(1)**     | **O(n)**   |
| `dict.get(key)`             | Safe key access (returns default if missing)        | **O(1)**     | **O(n)**   |
| `dict.setdefault(key, val)` | Get or insert key with default                      | **O(1)**     | **O(n)**   |

```python
# ===========================================
# 8 dict.popitem() â†’ Removes last inserted item
# ===========================================

# Sample dictionary
student ={"name":"Pankaj","age":27,"address":"Bhihar"}
print("Original dictionary:", student)

# âœ… popitem() removes and returns the last inserted (key, value) pair
last_pair = student.popitem()
print("\nRemoved item (key, value):", last_pair)

# âœ… Dictionary after removing the last item
print("Updated dictionary:", student)

# ðŸ§  Note:
#   - Before Python 3.7: popitem() removed a random item.
#   - From Python 3.7+: dicts preserve insertion order, so popitem() removes the *last inserted* item.

# âœ… If you call popitem() on an empty dictionary, it raises a KeyError.
empty_dict = {}
# empty_dict.popitem()  # âŒ KeyError
```

```python
# ===========================================
# 9 dict.setdefault(key, default)
# -------------------------------------------
# âž¤ If key exists â†’ returns its current value.
# âž¤ If key doesnâ€™t exist â†’ adds it with the given default value,
#   then returns that value.
# ===========================================

student = {
    "name": "Pankaj",
    "age": 22,
    "course": "AI"
}

print("Original dictionary:", student)

# Case 1: Key exists
existing = student.setdefault("name", "Unknown")
print("\nIf key exists ('name') â†’", existing)
print("Dictionary after Case 1:", student)

# Case 2: Key doesn't exist
new_value = student.setdefault("marks", 95)
print("\nIf key doesn't exist ('marks') â†’", new_value)
print("Dictionary after Case 2:", student)
```

```python
# ===========================================
# 10 dict.update(other_dict)
# -------------------------------------------
# âž¤ Adds or updates multiple keyâ€“value pairs at once.
# âž¤ If a key exists â†’ its value is updated.
# âž¤ If a key doesnâ€™t exist â†’ itâ€™s added.
# ===========================================

student = {
    "name": "Pankaj",
    "age": 22,
    "course": "AI"
}

print("Original dictionary:", student)

# Another dictionary to merge
extra_info = {
    "age": 23,              # existing key â†’ updated
    "city": "Jaipur",       # new key â†’ added
    "language": "Python"    # new key â†’ added
}

# âœ… Update student dictionary
student.update(extra_info)
print("\nAfter update() â†’", student)
```

### Dictionary Operation / built in Functions 

```python
# ===========================================
# 1ï¸âƒ£ 'in' and 'not in'
# -------------------------------------------
# âž¤ Checks if a key exists in the dictionary (not value!)
# âž¤ Average TC: O(1) because dict uses hashing
# ===========================================

student = {
    "name": "Pankaj",
    "age": 22,
    "marks": 95,
    "course": "AI"
}
print("Dictionary:", student)

print("'age' in student:", 'age' in student)         # True
print("'city' not in student:", 'city' not in student)  # True

# âš ï¸ Note: membership checks only test keys, not values
print("'Pankaj' in student:", 'Pankaj' in student)   # False
```

```python
# ===========================================
# 2ï¸âƒ£ len(dict)
# -------------------------------------------
# âž¤ Returns number of key-value pairs in dict.
# âž¤ Time Complexity: O(1)
# ===========================================

print("\nLength of dict:", len(student))  # 4
```

```python
# ===========================================
# 3ï¸âƒ£ all(dict)
# -------------------------------------------
# âž¤ Returns True if all keys are truthy (non-zero / non-empty).
# âž¤ Evaluates over keys by default.
# âž¤ Time Complexity: O(n)
# ===========================================

student = {
    "name": "Pankaj",
    "age": 22,
    "marks": 95,
    "course": "AI"
}
print("Dictionary:", student)

student_truthy = {"a": 1, "b": 2}
student_falsey = {"a": 1, "": 2}  # empty string is False

print("\nall(student_truthy):", all(student_truthy))    # True
print("all(student_falsey):", all(student_falsey))      # False
```

```python
# ===========================================
# 4ï¸âƒ£ any(dict)
# -------------------------------------------
# âž¤ Returns True if at least one key is truthy.
# âž¤ Evaluates over keys by default.
# âž¤ Time Complexity: O(n)
# ===========================================

student = {
    "name": "Pankaj",
    "age": 22,
    "marks": 95,
    "course": "AI"
}
print("Dictionary:", student)

empty_dict = {}
some_falsey = {"": 0, 0: None, False: 1}

print("\nany(student):", any(student))          # True â†’ has truthy keys
print("any(empty_dict):", any(empty_dict))      # False â†’ no keys
print("any(some_falsey):", any(some_falsey))    # False â†’ all keys falsy
```

```python
# ===========================================
# 5ï¸âƒ£ sorted(dict)
# -------------------------------------------
# âž¤ Returns a sorted list of dictionary keys.
# âž¤ Sorting compares keys (not values) by default.
# âž¤ Time Complexity: O(n log n)
# ===========================================
student = {
    "age": 22,
    "marks": 3,
}
print("Dictionary:", student)

print("\nSorted keys:", sorted(student))  # Alphabetical order of keys

# âœ… Sort by values using 'key' argument
print("Sorted by values:", sorted(student, key=lambda k: student[k]))
```

### Dictionary vs List

ðŸ“˜ **Dictionary vs List in Python â€” Quick Comparison**

| Feature                 | **Dictionary (`dict`)**               | **List (`list`)**                                    |
| ----------------------- | ------------------------------------- | ---------------------------------------------------- |
| **Structure**           | Collection of **keyâ€“value pairs**     | Ordered collection of **elements**                   |
| **Access Method**       | Access by **key** â†’ `dict[key]`       | Access by **index** â†’ `list[index]`                  |
| **Order (Python 3.7+)** | Preserves **insertion order**         | Always **ordered**                                   |
| **Duplicates**          | Keys must be **unique**               | Elements can **repeat**                              |
| **Mutability**          | âœ… Mutable (can add, remove, modify)   | âœ… Mutable                                            |
| **When to Use**         | When you need **fast lookups by key** | When you need **ordered sequences** or **iteration** |
| **Lookup Speed**        | **O(1)** (average, hash table based)  | **O(n)** (linear search)                             |
| **Insertion Speed**     | **O(1)** (amortized)                  | **O(1)** (append at end)                             |
| **Deletion Speed**      | **O(1)** (by key)                     | **O(n)** (by value/index)                            |
| **Memory Usage**        | Higher (stores keys + values + hash)  | Lower                                                |
| **Example Use Case**    | Storing student records `{id: name}`  | Storing list of marks `[89, 90, 92]`                 |

ðŸ§  **Summary:**

* Use a **list** when order and duplicates matter.
* Use a **dictionary** when you need **fast key-based access** and meaningful associations between items (like â€œname â†’ scoreâ€).


### Dictionary_Comprehension

```python
# ============================================================
# ðŸ“˜ Dictionary Comprehensions in Python (Colab-ready, 1 cell)
# From basic to advanced, with clear, runnable examples
# ============================================================

# ------------------------------------------------------------
# 0) Setup sample data used across examples
# ------------------------------------------------------------
nums = [1, 2, 3, 4, 5, 6]
names = ["alice", "bob", "charlie", "dave"]
scores = [91, 84, 77, 92]
pairs = [("a", 1), ("b", 2), ("c", 3), ("d", 4)]
sentence = "to be or not to be that is the question"
records = [
    {"id": 101, "name": "alice", "age": 23},
    {"id": 102, "name": "bob", "age": 21},
    {"id": 103, "name": "charlie", "age": 25},
]

print("Setup complete.\n")

# ------------------------------------------------------------
# 1) BASIC: transform a list into a dict  (key -> value)
# ------------------------------------------------------------
squares = {x: x * x for x in nums}
print("1) squares:", squares)

# ------------------------------------------------------------
# 2) FILTERING: include only items passing a condition
# ------------------------------------------------------------
even_squares = {x: x * x for x in nums if x % 2 == 0}
print("2) even_squares:", even_squares)

# ------------------------------------------------------------
# 3) CONDITIONAL VALUES: if/else inside the value position
# ------------------------------------------------------------
parity = {x: ("even" if x % 2 == 0 else "odd") for x in nums}
print("3) parity:", parity)

# ------------------------------------------------------------
# 4) FROM TWO LISTS: zip() to combine keys and values
#    (lists must be same length; zip stops at the shortest)
# ------------------------------------------------------------
name_to_score = {n: s for n, s in zip(names, scores)}
print("4) name_to_score:", name_to_score)

# ------------------------------------------------------------
# 5) FROM LIST OF TUPLES / PAIRS
# ------------------------------------------------------------
from_pairs = {k: v for k, v in pairs}
print("5) from_pairs:", from_pairs)

# ------------------------------------------------------------
# 6) USING ENUMERATE: build index -> value dict
# ------------------------------------------------------------
index_to_name = {i: n for i, n in enumerate(names)}
print("6) index_to_name:", index_to_name)

# ------------------------------------------------------------
# 7) TRANSFORM KEYS/VALUES: e.g., title-case names, curve scores
# ------------------------------------------------------------
pretty_scores = {n.title(): min(100, s + 5) for n, s in name_to_score.items()}
print("7) pretty_scores (curved + title names):", pretty_scores)

# ------------------------------------------------------------
# 8) FILTER BY KEY OR VALUE: keep only scores >= 90
# ------------------------------------------------------------
top_students = {n: s for n, s in name_to_score.items() if s >= 90}
print("8) top_students:", top_students)

# ------------------------------------------------------------
# 9) INVERT A DICT: value -> key
#    âš ï¸ Only safe if values are unique; otherwise later keys overwrite earlier ones
# ------------------------------------------------------------
score_to_name = {s: n for n, s in name_to_score.items()}
print("9) score_to_name (inverted):", score_to_name)

# ------------------------------------------------------------
# 10) COUNT FREQUENCIES: word -> count (comprehension over set of words)
#     We use set(...) so we compute once per unique word.
# ------------------------------------------------------------
words = sentence.split()
freq = {w: words.count(w) for w in set(words)}
print("10) freq:", freq)

# ------------------------------------------------------------
# 11) NESTED COMPREHENSION: table of (i, j) -> i*j (small example)
# ------------------------------------------------------------
table = {(i, j): i * j for i in range(1, 4) for j in range(1, 4)}
print("11) table (i,j)->i*j:", table)

# ------------------------------------------------------------
# 12) COMPREHENSION WITH COMPLEX CONDITION:
#     only include (name, score) where name starts with 'c' or 'd' and score >= 80
# ------------------------------------------------------------
filtered = {
    n: s
    for n, s in name_to_score.items()
    if n[0] in {"c", "d"} and s >= 80
}
print("12) filtered (c/d & >=80):", filtered)

# ------------------------------------------------------------
# 13) NORMALIZE / CLEAN DATA: strip spaces, upper-case keys
# ------------------------------------------------------------
raw = {"  key1  ": "  val1", " key2": "val2  ", "KEY3  ": "  val3  "}
cleaned = {k.strip().upper(): v.strip() for k, v in raw.items()}
print("13) cleaned:", cleaned)

# ------------------------------------------------------------
# 14) FROM LIST OF RECORDS: map id -> (name, age)
# ------------------------------------------------------------
id_to_person = {r["id"]: (r["name"], r["age"]) for r in records}
print("14) id_to_person:", id_to_person)

# ------------------------------------------------------------
# 15) GROUP-LIKE KEYING (without side-effects):
#     Build first-letter -> count (use sum over list comprehension)
#     (For full grouping to lists, use loops or defaultdict in practice.)
# ------------------------------------------------------------
first_letter_count = {
    ch: sum(1 for n in names if n.startswith(ch))
    for ch in {n[0] for n in names}
}
print("15) first_letter_count:", first_letter_count)

# ------------------------------------------------------------
# 16) DICT COMPREHENSION WITH FUNCTION APPLICATION:
#     apply a function to values during construction
# ------------------------------------------------------------
def grade(score):
    return "A" if score >= 90 else "B" if score >= 80 else "C"

grades = {n: grade(s) for n, s in name_to_score.items()}
print("16) grades:", grades)

# ------------------------------------------------------------
# 17) FLATTEN SELECTIVE FIELDS: (index -> selected field)
# ------------------------------------------------------------
idx_to_upper_name = {i: rec["name"].upper() for i, rec in enumerate(records)}
print("17) idx_to_upper_name:", idx_to_upper_name)

# ------------------------------------------------------------
# 18) MULTI-KEY BUILD: (name, length) -> score
# ------------------------------------------------------------
name_len_key = {(n, len(n)): s for n, s in name_to_score.items()}
print("18) name_len_key:", name_len_key)

# ------------------------------------------------------------
# 19) DEFAULT/SAFE LOOKUPS WHILE BUILDING:
#     Suppose some records may lack 'age'â€”use dict.get with a default.
# ------------------------------------------------------------
maybe_missing = [{"name": "eve"}, {"name": "frank", "age": 20}]
name_to_age = {r["name"]: r.get("age", -1) for r in maybe_missing}
print("19) name_to_age (get with default):", name_to_age)

# ------------------------------------------------------------
# 20) COMPREHENSION + SET/UNIQUE FILTER:
#     Build a dict from unique numbers: num -> num**3 for evens only
# ------------------------------------------------------------
unique_nums = {2, 2, 4, 6, 6, 8}
even_cubes = {x: x ** 3 for x in unique_nums if x % 2 == 0}
print("20) even_cubes from unique set:", even_cubes)

print("\nAll examples executed successfully âœ…")
```
